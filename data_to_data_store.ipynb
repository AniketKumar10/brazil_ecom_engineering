```python
import pandas as pd
from sqlalchemy import create_engine

# Define the path to your CSV file
csv_file_path = "dataset/olist_order_items_dataset.csv"

# Create a connection to the MySQL database
engine = create_engine('mysql+mysqlconnector://username:password@localhost/database_name')

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

# Write the DataFrame to a MySQL table
df.to_sql(name='order_items_dataset', con=engine, if_exists='replace', index=False)

print("Data imported successfully!")
```
